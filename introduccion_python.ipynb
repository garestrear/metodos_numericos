# **3. Tipos de datos**
Hay varios tipos de datos:
* **Int:** números enteros como $0$, $-34$, $26$.
* **Float:** son los números reales, es decir, los números enteros junto con los números decimales: $2/3$, $-9.45$, $5.0$
* **Booleanos:** son solo dos valores: Verdadero o Falso, pero siempre los escribiremos en inglés: `True` y `False` (a veces se entienden como 1 y 0 --por ejemplo, intenta ejecutar `True + True`). Hay que tener cuidado en escribirlos siempre con la primera letra en mayúscula.
* **String:** es decir, cadenas. Son los datos que escribimos siempre entre comillas doble `" "` ó simples `' '`. Aquí podemos usar letras, números o símbolos. Por ejemplo: `"Jorge"`, `"cinco"`, `"Python3.1"`, `"Aprenderemos mucho!!"`, `"5"`.

#  **10.** `for`

Empezamos definiendo un bucle o loop. La idea es hacer que una parte de nuestro código se ejecute varias veces para no tener que hacerlo nosotros. Para esto usamos un `for`.

# **11.** `if`

Definimos ahora un condicional. La idea es ponerle condiciones a ciertas partes de nuestro código para que dichas  partes del código se ejecuten o no.

**Tu turno:** En la siguiente celda de código usa el símbolo `#` para comentar lo que hace cada instrucción.


````
# Cambia las rayitas después del númeral por tus comentarios
notas = [3, 3.3, 2.3, 3.5, 2.7]     #--------
total_notas = len(notas)            #--------
suma = 0                            #--------
for i in notas:                     #--------
  suma = suma + i                   #--------
promedio = suma/total_notas         #--------
print(promedio)                     #--------
if promedio >= 3:                   #--------
  print('¡Soy genial, aprobé el semestre!')      #--------
else:                               #--------
  print('¡Huy, me tiré el semestre!')    #--------
````

#**13.** `while`
El bucle `while` ejecuta una acción **siempre y cuando** una condición se satisfaga. Veamos un ejemplo en la siguiente celda de código.


````
i = 1
while i <= 10:
  print(i)
  i = i + 1
````

Lo anterior también lo podríamos hacer con un bucle `for` y haciendo uso de la función `range()`. Sin embargo, el bucle `while` nos permite imponer condiciones más complicadas.

````
j = 3
while j > 0.1:
  print(j)
  j = j / 2 
````

````
#importamos el módulo time
import time

#Bienvenida para el usuario.
nombre = input("Cual es tu nombre? ")

print("Hola, " + nombre, "Vamos a jugar al ahorcado!")

print(" ")

#Hacemos una pausa de un segundo.
time.sleep(1)

print("Empecemos a adivinar...")
time.sleep(0.5)

#Aqui escondemos la palabra.
palabra = "python"

#creamos una variable con una cadena vacia
aciertos = ''

#determinamos el numero de turnos
turnos = 10

# Creamos un loop while

#revisamos que los turnos sean mayores que cero.
while turnos > 0:         

    # empezamos a contar el numero de fallas
    fallas = 0             

    # para toda letra en la palabra secreta   
    for letra in palabra:      

    # vemos si la letra esta dentro de las que el jugardor ya adivino
        if letra in aciertos:    
    
        # imprimimos la letra
            print(letra, end= '')    

        else:
    
        # si no ponemos un guion bajo.
            print(" _ ", end= '')
       
        # e incrementamos el numero de fallas
            fallas += 1    

    # si las fallas son iguales a cero

    # Imprimimos que el usuario gano.
    if fallas == 0:
        print(" ")      
        print("Ganaste! Sigue aprendiendo Python con nosotros!")

    # terminamos el programa
        break              

    print("")

    # Le pedimos al usuario adivinar una letra
    letra = input("Adivina una letra:") 

    # agregamos la letra a los aciertos
    aciertos += letra                    

    # si la letra no esta en la palabra secreta
    if letra not in palabra:  
 
     # hacemos que pierna un turno 1 (ahora 9)
        turnos -= 1        
 
    # imprimimos error
        print(" ")
        print("Letra incorrecta. Sigue intentándolo y verás que lo lograrás.")    
 
    # le mostramos al usuario cuantos turnos le quedan
        print("Te quedan",  turnos, 'turnos mas.')
 
    # si los turnos se acaban
        if turnos == 0:           
    
        # imprimimos que perdio.
            print(" ")
            print("Lo sentimos, se te acabaron los turnos. Pero no dejes de aprender Python!")
````
